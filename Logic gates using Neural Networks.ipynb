```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##define sigmoid function
def sigmoid(x):
    return 1.0/ (1.0+ np.exp(-x))
```

```python
##plot the signoid func

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show()
```

```python
def logic_gate(w1, w2, b):
    return lambda x1, x2: sigmoid(w1 * x1 + w2 * x2 + b)

def test(gate):
    for a, b in ((0, 0), (0, 1), (1, 0), (1, 1)):
        print("{}, {} : {}".format(a, b, np.round(gate(a, b))))
```

```python
print("AND Gate: ")
AND_gate = logic_gate(5, 5, -7)   # This approximates an AND gate
test(AND_gate)

print("OR Gate: ")
OR_gate = logic_gate(5, 5, -2)    # This approximates an OR gate
test(OR_gate)

print("NAND Gate: ")
NAND_gate = logic_gate(-5, -5, 7) # This approximates a NAND gate
test(NAND_gate)
```

```python
##XOR gate using 2 neural layer
def XOR_nn(x1, x2):
    # Hidden layer
    h1 = sigmoid(-5 * x1 + -5 * x2 + 7)  # NAND
    h2 = sigmoid(5 * x1 + 5 * x2 + -2)   # OR

    # Round hidden outputs to binary (simulate threshold)
    h1_bin = np.round(h1)
    h2_bin = np.round(h2)

    # Output layer (AND)
    y = sigmoid(5 * h1_bin + 5 * h2_bin + -7)
    return np.round(y)
```

```python
print("XOR Gate: ")
def test(gate):
    for a, b in ((0, 0), (0, 1), (1, 0), (1, 1)):
        print(f"{a}, {b} : {gate(a, b)}")

test(XOR_nn)
```
